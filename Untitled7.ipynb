```python
import numpy as np
import pandas as pd



# Extract the numerical columns for PCA
numerical_data = df1.select_dtypes(include=[np.number])

# Step 1: Standardization
mean = numerical_data.mean()
std_dev = numerical_data.std()
standardized_data = (numerical_data - mean) / std_dev

# Step 2: Covariance Matrix
cov_matrix = np.cov(standardized_data, rowvar=False)

# Step 3: Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 4: Sorting Eigenvalues
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Step 5: Percentage of Variance Explained
total_variance = np.sum(sorted_eigenvalues)
variance_explained_first = sorted_eigenvalues[0] / total_variance
variance_explained_second = (sorted_eigenvalues[0] + sorted_eigenvalues[1]) / total_variance

print(f"Percentage of Variance Explained by the First Principal Component: {variance_explained_first * 100:.2f}%")
print(f"Percentage of Variance Explained by the First and Second Principal Components: {variance_explained_second * 100:.2f}%")
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Sample data (replace with your dataset)
location_data = df1['Location'].values
operator_data = df1['Operator'].values
data = np.column_stack((location_data, operator_data))

# Number of clusters (K)
K = 3  # You can change this to the desired number of clusters

# Randomly initialize cluster centroids
np.random.seed(0)  # For reproducibility
centroids = data[np.random.choice(len(data), K, replace=False)]

# Maximum number of iterations (adjust as needed)
max_iterations = 100

# Custom K-means algorithm
for iteration in range(max_iterations):
    # Assign data points to clusters
    cluster_labels = []
    for point in data:
        distances = [math.sqrt(np.sum((point - centroid) ** 2)) for centroid in centroids]
        cluster_labels.append(np.argmin(distances))
    cluster_labels = np.array(cluster_labels)

    # Update centroids as the mean of data points in each cluster
    new_centroids = np.array([data[cluster_labels == k].mean(axis=0) for k in range(K)])

    # Check for convergence
    if np.all(centroids == new_centroids):
        break

    centroids = new_centroids

# Plot the results
plt.figure(figsize=(10, 6))
for k in range(K):
    cluster_data = data[cluster_labels == k]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f'Cluster {k + 1}')

plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker='x', s=100, label='Centroids')
plt.title('K-means Clustering for Location and Operator Columns')
plt.xlabel('Location')
plt.ylabel('Operator')
plt.legend()
plt.show()
```

```python
# Sample data (replace with your dataset)
data = df1['Type'].tolist()

# Number of clusters (K)
K = 3  # You can change this to the desired number of clusters

# Initialize K-means with the number of clusters
import random

# Randomly initialize cluster centroids
random.seed(0)
centroids = random.sample(data, K)

# Maximum number of iterations (adjust as needed)
max_iterations = 100

# Custom K-means algorithm
for iteration in range(max_iterations):
    # Assign data points to clusters
    cluster_assignments = []
    for x in data:
        closest_centroid = min(centroids, key=lambda c: abs(x - c))
        cluster_assignments.append(closest_centroid)

    # Update centroids as the mean of data points in each cluster
    new_centroids = [np.mean([x for x, cluster in zip(data, cluster_assignments) if cluster == centroid]) for centroid in centroids]

    # Check for convergence
    if centroids == new_centroids:
        break

    centroids = new_centroids

# Calculate silhouette score manually
n = len(data)
silhouette_values = []

for i in range(n):
    a_i = np.mean([abs(data[i] - data[j]) for j in range(n) if j != i and cluster_assignments[j] == cluster_assignments[i]])
    b_i = min([np.mean([abs(data[i] - data[j]) for j in range(n) if cluster_assignments[j] == k]) for k in centroids if k != cluster_assignments[i]])
    silhouette_i = (b_i - a_i) / max(a_i, b_i)
    silhouette_values.append(silhouette_i)

silhouette_avg = np.mean(silhouette_values)
print(f"Silhouette Score for K-means clustering on 'Type' Column: {silhouette_avg:.2f}")
```
