```python
import numpy as np

education = np.array(df['Education'])
experience = np.array(df['Experience (Years)'])
encoded_occupation = np.array(df['encoded_occupation'])
salary = np.array(df['Salary'])
```

```python
import numpy as np


# Perform multiple linear regression
X = np.column_stack((education, experience, encoded_occupation, np.ones(len(education))))
coefficients = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(salary)
beta_education, beta_experience, beta_encoded_occupation, beta_intercept = coefficients

print("Multiple Linear Regression Results:")
print(f"Beta Education: {beta_education:.4f}")
print(f"Beta Experience: {beta_experience:.4f}")
print(f"Beta Encoded Occupation: {beta_encoded_occupation:.4f}")
print(f"Beta Intercept: {beta_intercept:.4f}")

# Perform polynomial regression (degree 2 on 'education' variable)
degree = 2
X_poly = np.column_stack((education**i for i in range(1, degree + 1)))
X_poly = np.column_stack((X_poly, experience, encoded_occupation, np.ones(len(education))))
coefficients_poly = np.linalg.inv(X_poly.T.dot(X_poly)).dot(X_poly.T).dot(salary)
beta_education_poly = coefficients_poly[0]
beta_experience_poly = coefficients_poly[1]
beta_encoded_occupation_poly = coefficients_poly[2]
beta_intercept_poly = coefficients_poly[-1]

print("\nPolynomial Regression (Degree 2 on Education) Results:")
print(f"Beta Education (Poly): {beta_education_poly:.4f}")
print(f"Beta Experience (Poly): {beta_experience_poly:.4f}")
print(f"Beta Encoded Occupation (Poly): {beta_encoded_occupation_poly:.4f}")
print(f"Beta Intercept (Poly): {beta_intercept_poly:.4f}")
```

```python
# Stack the relevant features together
relevant_features = np.column_stack((education, experience, encoded_occupation))

# Perform PCA
mean_centered = relevant_features - relevant_features.mean(axis=0)
cov_matrix = np.cov(mean_centered, rowvar=False)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
explained_variance_ratio = eigenvalues / eigenvalues.sum()

# Choose the number of components (e.g., 2 for this example)
num_components = 2
top_eigenvectors = eigenvectors[:, :num_components]

# Transform the data using PCA
transformed_data = np.dot(mean_centered, top_eigenvectors)

# Perform linear regression on the transformed data
X_pca = np.column_stack((transformed_data, np.ones(len(transformed_data))))
coefficients_pca = np.linalg.inv(X_pca.T.dot(X_pca)).dot(X_pca.T).dot(salary)
beta_pca = coefficients_pca[:-1]
intercept_pca = coefficients_pca[-1]

print("PCA Regression Results:")
print(f"Beta PCA: {beta_pca}")
print(f"Intercept PCA: {intercept_pca}")
```

```python
# Function to perform Gram-Schmidt orthogonalization
def gram_schmidt(X):
    Q, R = np.linalg.qr(X)
    return Q

# Perform Gram-Schmidt orthogonalization
orthogonalized_features = gram_schmidt(relevant_features)

# Perform multiple linear regression on orthogonalized data
X_orthogonal = np.column_stack((orthogonalized_features, np.ones(len(orthogonalized_features))))
coefficients_orthogonal = np.linalg.inv(X_orthogonal.T.dot(X_orthogonal)).dot(X_orthogonal.T).dot(salary)
beta_orthogonal = coefficients_orthogonal[:-1]
intercept_orthogonal = coefficients_orthogonal[-1]

print("Orthogonalized Regression Results:")
print(f"Beta Orthogonal: {beta_orthogonal}")
print(f"Intercept Orthogonal: {intercept_orthogonal}")

# Plot original "Experience" vs. orthogonalized "Experience"
plt.scatter(experience, orthogonalized_features[:, 1])
plt.xlabel("Original Experience")
plt.ylabel("Orthogonalized Experience")
plt.title("Scatter Plot: Original vs. Orthogonalized Experience")
plt.grid(True)
plt.show()
```

```python
X =df.iloc[:,0:3]
ones = np.ones([X.shape[0],1])
X = np.concatenate((ones,X),axis=1)

y = df.iloc[:,-1].values
theta = np.zeros([1,3])

#set hyper parameters
alpha = 0.01
iters = 1000
```

```python
def computeCost(X,y,theta):
    tobesummed = np.power(((X @ theta.T)-y),2)
    return np.sum(tobesummed)/(2 * len(X))
```

```python
#gradient descent
def gradientDescent(X,y,theta,iters,alpha):
    cost = np.zeros(iters)
    for i in range(iters):
        theta = theta - (alpha/len(X)) * np.sum(X * (X @ theta.T - y), axis=0)
        cost[i] = computeCost(X, y, theta)

    return theta,cost

#running the gd and cost function
g,cost = gradientDescent(X,y,theta,iters,alpha)
print(g)

finalCost = computeCost(X,y,g)
print(finalCost)
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(iters), cost, 'r')
ax.set_xlabel('Iterations')
ax.set_ylabel('Cost')
ax.set_title('Error vs. Training Epoch')
```
